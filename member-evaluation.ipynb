```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('echantillon.xlsx')
df
```

```python
df.drop('Unnamed: 0', inplace=True, axis=1)
```

```python
df
```

```python
df['Points']=(1+0.02*df['Hors_Usthb']+0.03*df['Structure'])*(2*df['présence']+3*df['p_évenement']+4*df['o_évenement']+3*df['P_participation']+5*df['P_club']+2*df['easy']+3*df['moderate']+5*df['hard'])
```

```python
df
```

```python
leader_board = df.sort_values(by=['Points'], ascending=False)
leader_board
```

```python
rank = ['LEGEND']
l = len(leader_board)
```

```python
for i in range(int(0.05*l)):
    rank.append('Diamond')
for i in range(int(0.1*l)):
    rank.append('Platinium')
for i in range(int(0.15*l)):
    rank.append('Gold')
for i in range(int(0.2*l)):
    rank.append('Silver')
for i in range(int(0.25*l)):
    rank.append('Bronze')
for i in range(l-int(0.75*l)-1):
    rank.append('Rusty')
```

```python
leader_board['Rank'] = rank
leader_board
```

```python
leader_board.to_excel("output.xlsx")
```

```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_excel('ORSC_Members.xlsx')
```

```python
df1.drop('Unnamed: 0', inplace=True, axis=1)
df1
```

```python
df1['Points']=(1+0.04*df1['Structure']+0.02*df1['Hors_Usthb'])*(df1['p_réunion']+1.5*df1['idée_proposé']+1.5*df1['p_évenement']+2*df1['o_évenement']+2.5*df1['P_participation']+3*df1['P_proposé']+3.5*df1['P_club'])
```

```python
df1
```

```python
df = df1.sort_values(by=['Points'], ascending=False)
df
```

```python
rank = ['Legend']
l = len(df1)
```

```python
for i in range(int(0.05*l)):
    rank.append('Diamond')
for i in range(int(0.1*l)):
    rank.append('Platinium')
for i in range(int(0.15*l)):
    rank.append('Gold')
for i in range(int(0.2*l)):
    rank.append('Silver')
for i in range(int(0.25*l)):
    rank.append('Bronze')
for i in range(l-int(0.75*l)-1):
    rank.append('Rusty')
```

```python
df1['Rank']= rank
df1
```

```python
df1.to_excel("ORSC_Members_Evaluation.xlsx")  
```
